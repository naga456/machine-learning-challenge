```python
import pandas as pd
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head()
```

# Select your features (columns)

```python
# Set features. This will also be used as your x values.
#selected_features = df[['names', 'of', 'selected', 'features', 'here']]
X = df.drop("koi_disposition",axis=1)
y = df["koi_disposition"]
print(X.shape, y.shape)
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
X = X.drop(['ra', 'dec','koi_kepmag'], axis=1)
```

```python
data = y.copy()
data_binary_encoded = pd.get_dummies(data, columns = ["Koi_disposition"])
data_binary_encoded.head()
y =  data_binary_encoded
```

```python
y.head()
```

```python
y = y.drop(['CANDIDATE'], axis=1)

print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=1, stratify = y)
```

```python
X_train.head()
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# Scale your data
from sklearn.preprocessing import StandardScaler
X_scaler = StandardScaler().fit(X_train)
#y_scaler = StandardScaler().fit(y_train)

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
#y_train_scaled = y_scaler.transform(y_train)
#y_test_scaled = y_scaler.transform(y_test)
```

```python
print(X_train_scaled.shape,X_test_scaled.shape, y_train.shape,y_test.shape)
```

```python
y_train
```

```python
from sklearn.ensemble import RandomForestClassifier
model3 =RandomForestClassifier(n_estimators=50)
model3 = model3.fit(X_train_scaled,y_train)
```

# Train the Model



```python
print(f"Training Data Score: {model3.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {model3.score(X_test_scaled, y_test)}")
```

# Hyperparameter Tuning

Use `GridSearchCV` to tune the model's parameters

```python
# Create the GridSearchCV model
from sklearn.model_selection import GridSearchCV
param_grid = {'n_estimators': [50, 100, 200, 500],
              'max_depth': [10, 50, 100]}
grid3 = GridSearchCV(model3, param_grid, verbose=3)
```

```python
# Train the model with GridSearch
grid3.fit(X_train_scaled, y_train)
```

```python
print(grid3.best_params_)
print(grid3.best_score_)
```
